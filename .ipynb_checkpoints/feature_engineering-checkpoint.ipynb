```python
# phishing url ; count number of dots
def countdots(url):
    return url.count('.')

df['count_dots'] = df['url'].apply(lambda row: countdots(row))
```

```python
# phishing url ; count number of delimeters
def countdelim(url):
    count = 0
    delim = [';', '_', '?', '=', '&']
    for each in url:
        if each in delim:
            count = count + 1
    return count

df['count_delim'] = df['url'].apply(lambda row: countdelim(row))
```

```python
# ID addr check
import ipaddress as ip

def isip(url):
    try:
        if ip.ip_adress(url):
            return 1
    except:
        return 0
df['ip_check'] = df['url'].apply(lambda row: isip(row))
```

```python
# hyphens
def isPresentHyphen(url):
    return url.count('-')
df['count_hyphen'] = df['url'].apply(lambda row: isPresentHyphen(row))
```

```python
# check @
def isPresentAt(url):
    return url.count('@')
df['count_at'] = df['url'].apply(lambda row: isPresentAt(row))
```

```python
# check `//`
def isPresentDSlash(url):
    return url.count('//')
df['count_DSlach'] = df['url'].apply(lambda row: isPresentDSlash(row))
```

```python
# check `/`
def countSubDir(url):
    return url.count('/')
df['count_SubDir'] = df['url'].apply(lambda row: countSubDir(row))
```

```python
# Return the filename extension from url, or ''
import os
def get_ext(url):    
    root, ext = os.path.splitext(url)
    return ext
df['ext'] = df['url'].apply(lambda row: get_ext(row))
```

```python
def countSubDomain(subdomain):
    if not subdomain:
        return 0
    else:
        return len(subdomain.split('.'))
    
df['count_SubDomain'] = df['ext'].apply(lambda row: countSubDomain(row)) 
```

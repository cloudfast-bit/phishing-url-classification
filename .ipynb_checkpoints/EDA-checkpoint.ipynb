```python
# phishing url ; count number of dots
def countdots(url):
    return url.count('.')

df2['count_dots'] = df2['url'].apply(lambda row: countdots(row))
df2['count_dots_detail_url'] = df2['phish_detail_url'].apply(lambda row: countdots(row))
```

```python
# phishing url ; count number of delimeters
def countdelim(url):
    count = 0
    delim = [';', '_', '?', '=', '&']
    for each in url:
        if each in delim:
            count = count + 1
    return count

df2['count_delim'] = df2['url'].apply(lambda row: countdelim(row))
df2['count_delim_detail_url'] =df2['phish_detail_url'].apply(lambda row: countdelim(row))
```

```python
df2
```

```python
# ID addr check
import ipaddress as ip

def isip(url):
    try:
        if ip.ip_adress(url):
            return 1
    except:
        return 0
df2['ip_check'] = df2['url'].apply(lambda row: isip(row))
df2['ip_check_detail_url'] = df2['phish_detail_url'].apply(lambda row: isip(row))
```

```python
# hyphens
def isPresentHyphen(url):
    return url.count('-')
df2['count_hyphen'] = df2['url'].apply(lambda row: isPresentHyphen(row))
```

```python
# check @
def isPresentAt(url):
    return url.count('@')
df2['count_at'] = df2['url'].apply(lambda row: isPresentAt(row))
df2['count_at_detail_url'] = df2['phish_detail_url'].apply(lambda row: isPresentAt(row))
```

```python
# check `//`
def isPresentDSlash(url):
    return url.count('//')
df2['count_DSlach'] = df2['url'].apply(lambda row: isPresentDSlash(row))
df2['count_DSlach_detail_url'] = df2['phish_detail_url'].apply(lambda row: isPresentDSlash(row))
```

```python
# check `/`
def countSubDir(url):
    return url.count('/')
df2['count_SubDir'] = df2['url'].apply(lambda row: countSubDir(row))
df2['count_SubDir_detail_url'] = df2['phish_detail_url'].apply(lambda row: countSubDir(row))
```

```python
# Return the filename extension from url, or ''
import os
def get_ext(url):
    
    root, ext = os.path.splitext(url)
    return ext
df2['ext'] = df2['url'].apply(lambda row: get_ext(row))
df2['ext_detail_url'] = df2['phish_detail_url'].apply(lambda row: get_ext(row))
```

```python
df2
```

```python
def countSubDomain(subdomain):
    if not subdomain:
        return 0
    else:
        return len(subdomain.split('.'))
    
df2['count_SubDomain'] = df2['ext'].apply(lambda row: countSubDomain(row)) 
df2
```
